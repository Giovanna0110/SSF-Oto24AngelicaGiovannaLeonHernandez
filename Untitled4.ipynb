```python
# Tarea 5 . 1 implementación del ejemplo 4 . 1 . 3

import numpy as np
import matplotlib.pyplot as plt

# Parámetros del sistema
m = 1.0 # masa (kg)
g = 9.81 # aceleración de la gravedad (m/s^2)
k = 1.0 # constante del resorte (N/m)
x0 = 0.1 # posición inicial (m)
v0 = 0.0 # velocidad inicial (m/s)

# Definir el tiempo de simulación
t0 = 0.0 # tiempo inicial (s)
tf = 10.0 # tiempo final (s)
dt = 0.01 # paso de tiempo (s)

# Crear el arreglo de tiempo
t = np.arange(t0, tf + dt, dt)

# Definir las variables iniciales
x = np.zeros(len(t)) # posición
v = np.zeros(len(t)) # velocidad

# Establecer las condiciones iniciales
x[0] = x0
v[0] = v0

# Método de Euler para integrar la ecuación de movimiento
for i in range(1, len(t)):
    F = -m * g + k * x[i - 1] # fuerza en la masa
    a = F / m # aceleración
    v[i] = v[i - 1] + a * dt # velocidad
    x[i] = x[i - 1] + v[i] * dt # posición

# Generar la grafica del movimiento
plt.plot(t, x)
plt.title('Movimiento del péndulo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Posición (m)')
plt.grid(True)
plt.show()
```

```python
# Implementación del ejemplo 4 . 4 . 1  con el metodo de Runge-Kutta

import numpy as np
import matplotlib.pyplot as plt

# Función del método  de 2do orden
def rk2(y, time, dt, derivs):
    k0 = dt * derivs(y, time)
    k1 = dt * derivs(y + 0.5 * k0, time + 0.5 * dt)
    y_next = y + k1
    return y_next

# Derivadas para el sistema masa-resorte
def derivs(y, t, k):
    g = 9.8  # aceleración de la gravedad
    return np.array([y[1], -g + k * y[0]])

# Parámetros iniciales
y0 = np.array([0.1, 0.0])  # posición y velocidad iniciales
t0 = 0.0  # tiempo inicial
tf = 10.0  # tiempo final
dt = 0.01  # paso de tiempo

t = np.arange(t0, tf, dt)

# Valores de k a probar
k_values = [0.1, 0.2, 0.3, 1.0, 5.0, 10.0, 50.0, 100.0, 200.0]

# Simulación para cada valor de k
plt.figure(figsize=(10, 6))
for k in k_values:
    y = np.zeros((len(t), 2))
    y[0] = y0
    for i in range(1, len(t)):
        y[i] = rk2(y[i-1], t[i-1], dt, lambda y, t: derivs(y, t, k))

    # Graficar los resultados
    plt.plot(t, y[:, 0], label=f'k = {k}')

plt.title('Movimiento del péndulo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Posición (m)')
plt.legend()
plt.grid(True)
plt.show()
```
